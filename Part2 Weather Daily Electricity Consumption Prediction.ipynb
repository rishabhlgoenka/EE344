# Part 2: Weather -> Daily Electricity Consumption Prediction

**Dataset:** Electricity Consumption Based on Weather Data (Kaggle- https://www.kaggle.com/datasets/sudhirsingh27/electricity-consumption-based-on-weather-data)

**Task:** Build regression models to predict daily electricity consumption using weather features.

**Models:**
- Linear Regression
- Polynomial Regression (degree 2, 3, 4)

## 2.1 Load and Inspect the Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Load the dataset
df = pd.read_csv("data/electricity_consumption_based_weather_dataset.csv")

# Display column names and shape
print("Column Names:")
print(df.columns.tolist())
print("\nShape (rows, columns):", df.shape)
```

```python
# Display first few rows
print("First 5 rows:")
df.head()
```

```python
# Summary statistics
print("Summary Statistics:")
df.describe()
```

```python
# Check for missing values
print("Missing Values per Column:")
print(df.isna().sum())
print(f"\nTotal Missing Values: {df.isna().sum().sum()}")
```

```python
# Handle missing values by dropping rows with any missing data
df_clean = df.dropna()

print(f"Original shape: {df.shape}")
print(f"After dropping missing values: {df_clean.shape}")
print(f"Rows removed: {df.shape[0] - df_clean.shape[0]}")
```

### Dataset Summary

**Target (Dependent Variable):** `daily_consumption` — daily electricity consumption

**Features (Independent Variables):**
- `AWND` — Average wind speed
- `PRCP` — Precipitation
- `TMAX` — Maximum temperature
- `TMIN` — Minimum temperature

**Handling Missing Values:** Rows with missing values are dropped using `dropna()`.

## 2.2 Train/Test Split (70% / 30% Random)

```python
# Define features (X) and target (y)
# Features: weather variables (AWND, PRCP, TMAX, TMIN)
# Target: daily_consumption

X = df_clean[["AWND", "PRCP", "TMAX", "TMIN"]]
y = df_clean["daily_consumption"]

print("Features (X) shape:", X.shape)
print("Target (y) shape:", y.shape)
```

```python
# Split into train (70%) and test (30%)
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.30,
    random_state=42
)

print("Training set size:", len(X_train))
print("Testing set size:", len(X_test))
print(f"\nTrain/Test ratio: {len(X_train)/len(X)*100:.0f}% / {len(X_test)/len(X)*100:.0f}%")
```

## 2.3 Model Training: Linear + Polynomial Regression

1. **Linear Regression**
2. **Polynomial Regression (degree 2, 3, 4)**

No regularization used.

```python
# Model 1: Linear Regression
model_linear = LinearRegression()
model_linear.fit(X_train, y_train)

print("Linear Regression trained.")
```

```python
# Model 2: Polynomial Regression (Degree 2)
model_poly2 = Pipeline([
    ("poly", PolynomialFeatures(degree=2, include_bias=False)),
    ("lr", LinearRegression())
])
model_poly2.fit(X_train, y_train)

print("Polynomial Regression (degree=2) trained.")
```

```python
# Model 3: Polynomial Regression (Degree 3)
model_poly3 = Pipeline([
    ("poly", PolynomialFeatures(degree=3, include_bias=False)),
    ("lr", LinearRegression())
])
model_poly3.fit(X_train, y_train)

print("Polynomial Regression (degree=3) trained.")
```

```python
# Model 4: Polynomial Regression (Degree 4)
model_poly4 = Pipeline([
    ("poly", PolynomialFeatures(degree=4, include_bias=False)),
    ("lr", LinearRegression())
])
model_poly4.fit(X_train, y_train)

print("Polynomial Regression (degree=4) trained.")
```

## 2.4 Model Evaluation (Train and Test)

Metrics: **MSE**, **MAE**, **R²** on both train and test sets.

```python
# Helper function to compute metrics
def compute_metrics(y_true, y_pred):
    return {
        "MSE": mean_squared_error(y_true, y_pred),
        "MAE": mean_absolute_error(y_true, y_pred),
        "R^2": r2_score(y_true, y_pred),
    }
```

```python
# Evaluate all models
models = [
    ("Linear Regression", model_linear),
    ("Polynomial (degree=2)", model_poly2),
    ("Polynomial (degree=3)", model_poly3),
    ("Polynomial (degree=4)", model_poly4),
]

results = []
for model_name, model in models:
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    train_metrics = compute_metrics(y_train, y_train_pred)
    test_metrics = compute_metrics(y_test, y_test_pred)

    results.append({
        "Model": model_name,
        "Train MSE": train_metrics["MSE"],
        "Train MAE": train_metrics["MAE"],
        "Train R²": train_metrics["R^2"],
        "Test MSE": test_metrics["MSE"],
        "Test MAE": test_metrics["MAE"],
        "Test R²": test_metrics["R^2"],
    })
```

```python
# Display results table
results_df = pd.DataFrame(results)

print("=" * 80)
print("Model Comparison: Train vs Test Metrics")
print("=" * 80)
results_df
```

## 2.5 Discussion and Interpretation

### Which model generalizes best?

**Linear Regression** generalizes best with the highest Test R² (0.299) and lowest Test MSE (248,126).

This suggests the relationship between weather and electricity consumption is approximately linear — adding polynomial terms does not capture meaningful patterns, only noise.

### Do polynomial models improve the fit?

**No.** Polynomial models perform worse on the test set:
- Linear → Poly(2): Test R² decreased from 0.299 to 0.279
- Poly(2) → Poly(3): Test R² decreased from 0.279 to 0.250
- Poly(3) → Poly(4): Test R² **collapsed to -33.31** (catastrophic overfitting)

While Train R² increases with degree (0.276 → 0.296 → 0.311 → 0.330), this improvement doesn't generalize — the models are fitting training noise.

### Why does Polynomial (degree=4) fail so badly?

With 4 features and degree 4, `PolynomialFeatures` creates **many interaction terms** (70 features for degree 4). This leads to:
1. **Severe overfitting:** The model memorizes training data patterns that don't exist in test data
2. **Unstable coefficients:** Too many parameters for the signal available
3. Evidence: Train R² = 0.330 but Test R² = -33.31 (worse than predicting the mean)

### Why is overall model performance low (R² ≈ 0.30)?

All models achieve relatively low R² (~0.25-0.30). Two likely reasons:

1. **Limited feature set:** Weather alone (wind, precipitation, temperature) doesn't fully explain electricity consumption. Missing factors include:
   - Day of week / holidays (behavioral patterns)
   - Occupancy and human activity
   - Time of year / seasonal effects beyond temperature

2. **High noise / variability:** Daily consumption has significant variation (std = 607) not explained by weather, likely driven by unpredictable human behavior.

### Summary

| Finding | Evidence |
|---------|----------|
| Best model: Linear Regression | Highest Test R² (0.299) |
| Polynomial models hurt performance | Test R² decreases with higher degree |
| Degree 4 catastrophically overfits | Test R² = -33.31 |
| Weather explains ~30% of variance | All models R² ≈ 0.25-0.30 |

**Conclusion:** For this dataset, Linear Regression is the best choice. The weather features provide limited predictive power for electricity consumption — additional features (time, occupancy, etc.) would likely be needed to improve model performance.
